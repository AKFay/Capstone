```python
import pandas as pd
import numpy as np
from fastprogress.fastprogress import progress_bar
```

## Import data files

```python
file1 = pd.read_csv('GDSC1_public_raw_data_24Jul22.csv')
file1
```

```python
file2 = pd.read_csv('GDSC2_public_raw_data_24Jul22.csv')
file2
```

```python
# combine dataset 1 and dataset 2 into one big file

file_combo = pd.concat([file1, file2])
file_combo.drop_duplicates(inplace=True)
file_combo
```

## Data cleaning

```python
# figure out which conditions have 3 or more replicates

# unique plate, cell, tag, drug, and concentrations
replicates = pd.DataFrame(file_combo[['DRUGSET_ID','CELL_ID','TAG','DRUG_ID','CONC']].value_counts(dropna=False))
replicates = replicates.reset_index()
replicates.rename(columns={0:'replicates'}, inplace=True)

# remove conc column and drop duplicates so it's essentially only counting replicates for one conc
replicates.drop(['CONC'], axis=1, inplace=True)
replicates.drop_duplicates(inplace=True)

# only keep 3 or more replicates
replicates = replicates[replicates['replicates']>2]
replicates
```

```python
# remove samples we don't need

replicates = replicates[replicates['TAG']!='DMSO']
replicates = replicates[replicates['TAG']!='B']
replicates = replicates[replicates['TAG']!='UN-USED']
replicates = replicates[replicates['TAG']!='NC-1']
replicates
```

```python
# samples with NC-0 control

replicates_control  = replicates[replicates['TAG']=='NC-0']
replicates_control
```

```python
# make a list for indexing the file_combo and dropping rows we don't need

rep_ind = []

for i in replicates.index:
    index = replicates.loc[i]
    rep_ind.append((index['DRUGSET_ID'], index['CELL_ID'], index['DRUG_ID'], index['TAG']))
    
len(rep_ind)
```

```python
# make a new file that just has the data in replicates (by plate, cell line, drug, and tag)

rep_file = file_combo.copy()
rep_file.set_index(['DRUGSET_ID','CELL_ID','DRUG_ID','TAG'], inplace=True)
rep_file = rep_file.loc[rep_ind]
rep_file
```

```python
# double-check that everything has replicates

rep_df = pd.DataFrame(rep_file[['CELL_ID','DRUG_ID','TAG','CONC']].value_counts(dropna=False))
rep_df
```

```python
# number of curves to be generated

rep_df.reset_index(inplace=True)
rep_df = rep_df[rep_df['TAG']!='NC-0']
rep_df.drop(['CONC'], axis=1, inplace=True)
rep_df.drop_duplicates(inplace=True)
len(rep_df)
```

```python
rep_file.to_csv('rep_combo_file.csv')
```

## Data normalization

```python
# make a small test file for normalization

rep_file.reset_index(inplace=True)
test = rep_file[rep_file['DRUGSET_ID']=='713_a_17']
test
```

```python
# double-check that everything has replicates

pd.DataFrame(test[['CELL_ID','DRUG_ID','TAG','CONC']].value_counts(dropna=False))
```

```python
# make a new dataframe with a normalized intensity column
# 'list_comprehension' version

test['norm_intensity'] = [(test.loc[row]['INTENSITY']) / (test[(test['CELL_ID']==test.loc[row]['CELL_ID']) & (test['DRUGSET_ID']==test.loc[row]['DRUGSET_ID']) & (test['TAG']=='NC-0')]['INTENSITY'].copy().mean()) for row in progress_bar(test.index)]

# 'for loop' version
norm = pd.DataFrame()

for plate in progress_bar(test['DRUGSET_ID'].unique()):
    df = test[test['DRUGSET_ID']==plate].copy()
    for cell in df['CELL_ID'].unique():
        df2 = df[df['CELL_ID']==cell].copy()
        df2['norm_intensity_v2'] = df2['INTENSITY'] / df2[df2['TAG']=='NC-0']['INTENSITY'].mean()
        norm = pd.concat([norm, df2])
        
norm
```

```python
# 'for loop' version was much faster!
```

```python
test.to_csv('test_norm.csv',index=False)
```

## Limit to only triplicates

```python
# find all of the conditions with exactly 3 replicates

triplicates = pd.DataFrame(data_norm[['DRUGSET_ID','CELL_ID','DRUG_ID','TAG','CONC']].value_counts(dropna=False))
triplicates.reset_index(inplace=True)
triplicates.drop(['CONC'], axis=1, inplace=True)
triplicates.drop_duplicates(inplace=True)
triplicates = triplicates[triplicates[0]==3]
triplicates
```

```python
trip_ind = []

for i in triplicates.index:
    index = triplicates.loc[i]
    trip_ind.append((index['DRUGSET_ID'], index['CELL_ID'], index['DRUG_ID'], index['TAG']))
    
len(trip_ind)
```

```python
trip_file = data_norm.copy()
trip_file.set_index(['DRUGSET_ID','CELL_ID','DRUG_ID','TAG'], inplace=True)
trip_file = trip_file.loc[trip_ind]
trip_file
```
